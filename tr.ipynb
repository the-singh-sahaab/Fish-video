```python
import zipfile
import os
```

```python
zip_path = input("Enter the path to your zip file (e.g., /content/drive/MyDrive/yourfile.zip): ")

# Check if file exists
if not os.path.exists(zip_path):
    print(f"Error: File not found at {zip_path}")
else:
    # Get the folder name from zip file name (without extension)
    zip_name = os.path.basename(zip_path)
    folder_name = os.path.splitext(zip_name)[0]

    # Create extraction path in the same directory as the zip file
    zip_directory = os.path.dirname(zip_path)
    extract_path = os.path.join(zip_directory, folder_name)

    # Create the folder if it doesn't exist
    os.makedirs(extract_path, exist_ok=True)

    # Extract the zip file
    print(f"Extracting {zip_name} to {extract_path}...")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)

    print(f"✅ Successfully extracted to: {extract_path}")
    print(f"📁 Files extracted: {len(os.listdir(extract_path))} items")
```

```python
import cv2
import os

video_path = r'C:\Users\Shubh\Desktop\workspace\WhatsApp Video 2025-10-09 at 22.07.27_45b437c9.mp4'
output_folder = r'C:\Users\Shubh\Desktop\workspace\fish_image'
os.makedirs(output_folder, exist_ok=True)

cap = cv2.VideoCapture(video_path)
frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_path = os.path.join(output_folder, f"frame_{frame_count:06d}.jpg")
    cv2.imwrite(frame_path, frame)
    frame_count += 1

cap.release()
print(f"Extracted {frame_count} frames")
```

```python
import os

images_folder = r'C:\Users\Shubh\Desktop\workspace\fish_project1\images\train'
labels_folder = r'C:\Users\Shubh\Desktop\workspace\fish_project1\labels\train'

# Make sure the labels folder exists
os.makedirs(labels_folder, exist_ok=True)

# List all image files in the images folder
image_files = [f for f in os.listdir(images_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

for image_file in image_files:
    # Get the base name without extension
    base_name = os.path.splitext(image_file)[0]
    label_file = base_name + '.txt'
    label_path = os.path.join(labels_folder, label_file)

    # Check if this label file exists
    if not os.path.exists(label_path):
        # Create an empty label file
        open(label_path, 'w').close()
        print(f"Created empty label file: {label_path}")
    else:
        print(f"Label file already exists: {label_path}")
```

```python
import cv2
from ultralytics import YOLO
import os

# Load your trained model
model = YOLO(r"/content/runs/detect/train2/weights/best.pt")

# Input and output video paths
video_path = r"/content/drive/MyDrive/Pond1.mp4"
output_path = r"/content/drive/MyDrive/Pond1_ANS.mp4"

# Capture the video
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Ensure output directory exists
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# Initialize VideoWriter
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Run YOLO inference
    results = model(frame)
    annotated_frame = results[0].plot()

    # Write the annotated frame to output
    out.write(annotated_frame)

cap.release()
out.release()
print(f"Annotated video saved at: {output_path}")
```
